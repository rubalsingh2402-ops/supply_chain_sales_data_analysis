```python
df.columns= df.columns.str.strip().str.lower().str.replace(" ", "_")
df = df.rename(columns={
    '_salesteamid': 'sales_team_id',
    '_customerid': 'customer_id',
    '_storeid': 'store_id',
    '_productid': 'product_id'
})
df.columns
```

```python
id_cols = ['sales_team_id', 'customer_id', 'store_id', 'product_id']

for col in id_cols:
    if col in df.columns:
        df[col]=df[col].astype(str)
```

```python
date_cols = ['orderdate', 'shipdate', 'deliverydate', 'procureddate']

df[date_cols] = df[date_cols].apply(lambda x: pd.to_datetime(x, dayfirst=True, errors='coerce'))
```

```python
for col in ['unit_cost', 'unit_price'] : 
    df[col] = (
        df[col]
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False).astype(float))
```
